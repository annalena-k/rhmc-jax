```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt

from rhmcjax.targets.gaussians_on_circle import GaussiansOnCircle
```

```python
target = GaussiansOnCircle()
# Sampling
key = jax.random.key(3)
key, subkey = jax.random.split(key)
samples = target.sample(subkey, 1000)
fig, axs = plt.subplots(1,2, figsize=(6,3))
axs[0].scatter(samples[:,0], samples[:,1], s=5)
# Density Evaluation
x = jnp.linspace(-0.01, 1.01, 100)[1:-1]
grid_unit = jnp.dstack(jnp.meshgrid(x, x))
grid_unit = grid_unit.reshape([-1, 2])
density_vals = target.prob(grid_unit)
axs[1].scatter(grid_unit[:,0], grid_unit[:,1], c=density_vals);
```
